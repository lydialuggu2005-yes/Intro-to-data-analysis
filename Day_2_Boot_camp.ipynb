```python
cafe_order['quantity'] = pd.to_numeric(
    cafe_order['quantity'],
    errors = 'coerce'

)
```

```python
# IQR method — robust, no assumptions about distribution
q1, q3 = cafe_order["quantity"].quantile([0.25, 0.75])
iqr    = q3 - q1
lo, hi = q1 - 1.5*iqr,  q3 + 1.5*iqr

flagged =  cafe_order[(cafe_order["quantity"] < lo) |
             (cafe_order["quantity"] > hi)]
cafe_order["quantity"] = cafe_order["quantity"].clip(lo, hi)
```
